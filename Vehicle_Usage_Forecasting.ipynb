# VEHICLE USAGE FORECASTIN

 CSV must contain ONLY these columns:
 year | month | vehicle_type | vehicle_usage_count

 Logic:
 - Year growth learned from total yearly usage
 - Month growth learned from total monthly usage
 - Vehicle types are independent
 - No month-to-month dependency

Import All modules

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
```

-Import and Load CSV
-Check All required filed are present in the CSV or not if not through error

```python
CSV_PATH = "/content/Vehicle_Type_Forcas.csv"

df = pd.read_csv(CSV_PATH)

required_cols = {"year", "month", "vehicle_type", "vehicle_usage_count"}
assert required_cols.issubset(df.columns), "CSV columns are incorrect"
```

ENCODE VEHICLE TYPE (MODEL NEEDS NUMBERS)

```python
encoder = LabelEncoder()
df["vehicle_type_encoded"] = encoder.fit_transform(df["vehicle_type"])
```

Inputs → year, month, vehicle_type

```python
X = df[["year", "month", "vehicle_type_encoded"]]
```

Output → vehicle usage count

```python
y = df["vehicle_usage_count"]
```

 Train the Global Ensembel Model
 Use Decission Tree for Wisdom For Crowd

```python

model = RandomForestRegressor(
    n_estimators=300,
    max_depth=7,
    random_state=42
)

model.fit(X, y)
```

- Take Input From the User
- Prepare the input to fit the model for prediction

Feed the data input data to the model for predict the forecast and store it in a viable

Prepare the output and print the result

```python
future_year = int(input("Enter the year: "))
future_month = int(input("Enter the month: "))

future_rows = []
for v in encoder.classes_:
    future_rows.append([
        future_year,
        future_month,
        encoder.transform([v])[0]
    ])
```

```python
X_future = pd.DataFrame(
    future_rows,
    columns=["year", "month", "vehicle_type_encoded"]
)
predictions = model.predict(X_future)
```

```python
prediction_df = pd.DataFrame({
    "vehicle_type": encoder.classes_,
    "predicted_usage": predictions.astype(int)
})

print("=== Predicted Vehicle Usage ===")
print(prediction_df)
print("TOTAL VEHICLE USAGE:", prediction_df["predicted_usage"].sum())
```
